```python
from gbdxtools import Interface
gbdx = Interface()

filters = [ "platformName = WORLDVIEW03_VNIR" ]
try:
    results = gbdx.catalog.search(startDate="2019-05-21T00:00:00.000Z",endDate="2019-05-21T23:59:59.999Z",filters=filters)
    cat_ids = [r["identifier"] for r in results]
except Exception:
    cat_ids = []
len(cat_ids)
```

```python
from xml.etree import ElementTree
from xml.dom import minidom

def prettify(elem):
    rough_string = ElementTree.tostring(elem, 'utf-8')
    reparsed = minidom.parseString(rough_string)
    return reparsed.toprettyxml(indent="    ")
```

```python
from xml.etree.ElementTree import Element, SubElement
kml = Element('kml')
kml.set('xmlns', 'http://www.opengis.net/kml/2.2')
kml.set('xmlns:gx', 'http://www.google.com/kml/ext/2.2')
kml.set('xmlns:kml', 'http://www.opengis.net/kml/2.2')
kml.set('xmlns:atom', 'http://www.w3.org/2005/Atom')
doc = SubElement(kml, 'Document')
name1 = SubElement(doc, 'name')
name1.text = "WV03_VNIR Imagery from 2019-05-21T00:00:00.000Z to 2019-05-21T23:59:59.999Z"
open1 = SubElement(doc, 'open')
open1.text = '1'
style1 = SubElement(doc, 'Style')
style1.set('id', 'WV03_NOBS_NOMINAL')
line_style = SubElement(style1, 'LineStyle')
color1 = SubElement(line_style, 'color')
color1.text = 'ffffff00'
width1 = SubElement(line_style, 'width')
width1.text = '2'
poly_style = SubElement(style1, 'PolyStyle')
color2 = SubElement(poly_style, 'color')
color2.text = '40ffff00'
folder1 = SubElement(doc, 'Folder')
name2 = SubElement(folder1, 'name')
name2.text = "WV03"
open2 = SubElement(folder1, 'open')
open2.text = '1'
folder2 = SubElement(folder1, 'Folder')
name3 = SubElement(folder2, 'name')
name3.text = "NOBS"
open3 = SubElement(folder2, 'open')
open3.text = '1'
folder3 = SubElement(folder2, 'Folder')
name4 = SubElement(folder3, 'name')
name4.text = "NOMINAL"
#for i in range(0, len(results)):
for i in range(0, 5):
    place1 = SubElement(folder3, 'Placemark')
    pname1 = SubElement(place1, 'name')
    pname1.text = results[i]["identifier"]
    tz1 = SubElement(place1, 'TimeSpan')
    begin1 = SubElement(tz1, 'begin')
    begin1.text = results[i]["properties"]["timestamp"]
    end1 = SubElement(tz1, 'end')
    end1.text = results[i]["properties"]["timestamp"]
    sUrl1 = SubElement(place1, 'styleUrl')
    sUrl1.text = "#WV03_NOBS_NOMINAL"
    more_data1 = SubElement(place1, 'ExtendedData')
    prop1 = SubElement(more_data1, 'Data')
    prop1.set('name', 'catalogID')
    value1 = SubElement(prop1, 'value')
    value1.text = results[i]["identifier"]
    prop2 = SubElement(more_data1, 'Data')
    prop2.set('name', 'imageBands')
    value2 = SubElement(prop2, 'value')
    value2.text = results[i]["properties"]["imageBands"]
    prop3 = SubElement(more_data1, 'Data')
    prop3.set('name', 'sensorPlatformName')
    value3 = SubElement(prop3, 'value')
    value3.text = results[i]["properties"]["sensorPlatformName"]
    prop4 = SubElement(more_data1, 'Data')
    prop4.set('name', 'vendor')
    value4 = SubElement(prop4, 'value')
    value4.text = results[i]["properties"]["vendor"]
    prop5 = SubElement(more_data1, 'Data')
    prop5.set('name', 'cloudCover')
    value5 = SubElement(prop5, 'value')
    value5.text = str(results[i]["properties"]["cloudCover"])
    prop6 = SubElement(more_data1, 'Data')
    prop6.set('name', 'offNadirAngle')
    value6 = SubElement(prop6, 'value')
    value6.text = str(results[i]["properties"]["offNadirAngle"])
    prop8 = SubElement(more_data1, 'Data')
    prop8.set('name', 'sunElevation')
    value8 = SubElement(prop8, 'value')
    value8.text = str(results[i]["properties"]["sunElevation"])
    prop9 = SubElement(more_data1, 'Data')
    prop9.set('name', 'targetAzimuth')
    value9 = SubElement(prop9, 'value')
    value9.text = str(results[i]["properties"]["targetAzimuth"])
    prop10 = SubElement(more_data1, 'Data')
    prop10.set('name', 'browseURL')
    value10 = SubElement(prop10, 'value')
    value10.text = results[i]["properties"]["browseURL"]
    prop11 = SubElement(more_data1, 'Data')
    prop11.set('name', 'multiResolution')
    value11 = SubElement(prop11, 'value')
    value11.text = str(results[i]["properties"]["multiResolution"])
    prop12 = SubElement(more_data1, 'Data')
    prop12.set('name', 'panResolution')
    value12 = SubElement(prop12, 'value')
    value12.text = str(results[i]["properties"]["panResolution"])
    polygon1 = SubElement(place1, 'Polygon')
    outBound1 = SubElement(polygon1, 'outerBoundaryIs')
    linearRing1 = SubElement(outBound1, 'LinearRing')
    coord1 = SubElement(linearRing1, 'coordinates')
    prop1 = results[i]["properties"]["footprintWkt"]
    wkt1 = prop1[(prop1.find('(((') + 3):(prop1.find(')))'))]
    wkt1 = wkt1.replace(', ', ',0x')
    wkt1 = wkt1.replace(' ', ',')
    wkt1 = wkt1.replace(',0x', ',0 ')
    coord1.text = wkt1 + ',0'
    
print(prettify(kml))
# don't forget to change first line to <?xml version="1.0" encoding="UTF-8" ?>
# DO THIS before viewing in Google Earth!!!
```
